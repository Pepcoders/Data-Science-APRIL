```python
from sklearn.svm import LinearSVC
```

```python
x1 = np.arange(50)
x2 = np.random.rand(50)*50
X = np.dstack([x1, x2])[0]

plt.scatter(x1, x2)
```

```python
y = np.where(x1+x2 <= 45, 0, 1)

plt.scatter(x1, x2, c = y)
```

```python
model = LinearSVC().fit(X, y)

plot_decision_regions(X, y, clf = model)
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a two-dimensional SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
    levels=[-1, 0, 1], alpha=0.5,
    linestyles=['--', '-', '--'])
    # plot support vectors
    # if plot_support:
    #     ax.scatter(model.support_vectors_[:, 0],
    #     model.support_vectors_[:, 1],
    #     s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
model = LinearSVC(C=20).fit(X, y)
plt.scatter(x1, x2, c = y)

plot_svc_decision_function(model)
```

```python
from sklearn.datasets import make_circles

X, y = make_circles(noise=0.03)
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
df = pd.DataFrame(X, columns = ['x1', 'x2'])
df['y'] = y
df
```

```python
from sklearn.svm import LinearSVC

X = df.drop(columns=['y'])
y = df['y']

svc = LinearSVC().fit(X, y)
```

```python
plot_decision_regions(np.array(X), np.array(y), clf = svc)
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(noise=0.15, n_samples=500)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# gamma=0 underfit
# gamma=inf overfit

# C=0 underfit
# C=inf overfit

svc = SVC(kernel = 'rbf', gamma=100).fit(X, y)
# svc = SVC(kernel = 'poly', degree=5, C=50).fit(X, y)

plot_decision_regions(X, y, clf = svc)
```
